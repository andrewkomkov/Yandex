#### С такой таблицей уже можно работать

##### Создадим в нашей таблице столбцы income - в которых согласно наших данных будет расчитано сколько денежек приносит нам каждый из пользователей каждый месяц

**По интернет траффику условимся так - мегабайты сверх пакета считаются по мегабайтно как и в тарифе - не нужно покупать пакет из 1 ГБ что бы продолжить пользоваться интернетом**

```python
for i in range(1,13):
    prepaid_messages = bigdata['messages_included']
    prepaid_minutes = bigdata['minutes_included']
    prepaid_mb = bigdata['mg_per_month_included']
    rub_monthly_fee = bigdata['rub_monthly_fee']
    rub_per_message = bigdata['rub_per_message']
    rub_per_minute = bigdata['rub_per_minute']
    rub_per_gb = bigdata['rub_per_gb']
    messages_used = bigdata['messages_count_'+str(i)]
    minutes_used = bigdata['duration_'+str(i)]
    
    mb_used = bigdata['mb_used_'+str(i)]
    
    pay_for_minutes = (minutes_used - prepaid_minutes) 
    pay_for_messages = (messages_used - prepaid_messages) 
    
    pay_for_mb = (mb_used - prepaid_mb) 
    
    pay_for_minutes = pay_for_minutes.clip(lower=0)
    pay_for_messages = pay_for_messages.clip(lower=0)
    pay_for_mb = pay_for_mb.clip(lower=0)
    bigdata['income'+str(i)] = (pay_for_minutes*rub_per_minute)+(pay_for_messages*rub_per_message)+((pay_for_mb*(rub_per_gb/1024))).apply(np.ceil)+rub_monthly_fee
```

### Проведем статистический анализ данных:
- Создадим выборки
- Определим дисперсию и стандартное отклонение
- Построим гистограммы
- Определим и докажем(или опровергнем) наши гипотезы

```python
ultra_money = []
smart_money = []

def ultra_appender(g):
    if g > 0:
        ultra_money.append(g)

def smart_appender(g):
    if g > 0:
        smart_money.append(g)

bigdata.query('tariff == "ultra"')[income_columns].applymap(ultra_appender)
bigdata.query('tariff == "smart"')[income_columns].applymap(smart_appender)
print('Генеральные выборки сделаны')
print('Количество элементов в выборке по тарифу "Ультра":',len(ultra_money))
print('Количество элементов в выборке по тарифу "Смарт":',len(smart_money))
```

```python
print('Дисперсия элементов в выборке по тарифу "Смарт" равна',np.var(smart_money,ddof = 1))
print('Дисперсия элементов в выборке по тарифу "Ультра" равна',np.var(ultra_money,ddof = 1))

print('Стандартное отклонение элементов в выборке по тарифу "Смарт" равно',np.std(smart_money))
print('Стандартное отклонение элементов в выборке по тарифу "Ультра" равно',np.std(ultra_money))
```

#### Построим плотностные гистограммы

```python
bins = [500,700,900,1500,1700,2100,2300,2500,3000,3200,3500,5000,6000]

pd.Series(ultra_money).hist(bins = bins,alpha = 0.7)
pd.Series(smart_money).hist(bins = bins,alpha = 0.5)
```

Получается что оба наши датасета скошены вправо - оно и не удивительно, ведь начальные значения это стоимость базовая стоимость тарифа.

Еще можно сказать что пользователи тарифа "Ультра" более рационально подошли к выбору тарифа - очень не многие выходят за порог предоплаченных данных, в отличии от пользователей на тарифе "Смарт"

#### Посчитаем средние значения по тарифам:

```python
print('Среднее по тарифу Ультра:',sum(ultra_money)/len(ultra_money))
print('Среднее по тарифу Смарт:',sum(smart_money)/len(smart_money))
```

#### На первый взгляд средние траты пользователя Ультра куда выше, проверим это проведя т-тест

Сформируем гипотезы:
- Нулевая - Средния выручка пользователей обоих тарифов равна
- Альтернативная - выручка отличается

Значение альфы возьмём в 5 %

```python
alpha = 0.05

results = st.ttest_ind(
    ultra_money, 
    smart_money)

print('p-значение:',results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Средняя выручка у пользователей разных тарифов действительно различается и судя по величине pvalue - значительно

#### Проверим гипотезу о том что средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

```python
moscow_money = []
region_money = []

def moscow_appender(g):
    if g > 0:
        moscow_money.append(g)

def region_appender(g):
    if g > 0:
        region_money.append(g)

bigdata.query('city == "Москва"')[income_columns].applymap(moscow_appender)
bigdata.query('city != "Москва"')[income_columns].applymap(region_appender)
print('Генеральные выборки сделаны')
print('Количество элементов в выборке по городу Москва:',len(moscow_money))
print('Количество элементов в выборке по регионам:',len(region_money))
```

```python
print('Дисперсия элементов в выборке по городу Москва',np.var(moscow_money,ddof = 1))
print('Дисперсия элементов в выборке по регионам равна',np.var(region_money,ddof = 1))

print('Стандартное отклонение элементов в выборке по городу Москва равно',np.std(moscow_money))
print('Стандартное отклонение элементов в выборке по регионам равно',np.std(region_money))
```

```python
bins = [500,700,900,1500,1700,2100,2300,2500,3000,3200,3500,5000,6000]

pd.Series(moscow_money).hist(bins = bins,alpha = 0.7)
pd.Series(region_money).hist(bins = bins,alpha = 0.5)
```

По гистограмме видно что поведение пользователей не зависит от того - в столице они находятся или нет

Проведём т-тест для того чтобы подтвердить или опровергнуть это

```python
results = st.ttest_ind(
    moscow_money, 
    region_money)

print('p-значение:',results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Пользователи столицы тратят на мобильную связь столько же как и жители остальных регионов
